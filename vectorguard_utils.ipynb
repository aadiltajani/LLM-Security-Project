```python
import re
import json

# Open the log file
with open('exps_llama2_7b.log', 'r') as file:
    # Read the content of the file
    log_content = file.read()

# Use regular expressions to find all occurrences of the text between "Original message" and "it=1"
matches = re.findall(r'Original message:(.*?)it=1', log_content, re.DOTALL)

# Create a dictionary to store the extracted text
extracted_texts = {}

# Check if any matches are found
if matches:
    # Iterate over each match and store the extracted text in the dictionary
    for i, match in enumerate(matches, start=1):
        extracted_text = match.strip()
        extracted_texts[i] = extracted_text

# Write the extracted texts to a JSON file
with open('extracted_template_prompts.json', 'w') as json_file:
    json.dump(extracted_texts, json_file, indent=4)

print("Extraction completed. Extracted texts are stored in 'extracted_texts.json' file.")
```
